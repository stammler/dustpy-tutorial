# 1.1 Installation Test

Run this notebook to test if the installation of DustPy and DustPyLib has worked properly and the latest versions of the required packages are installed.

```python
import dustpy
import dustpylib
from IPython.display import Markdown as md
import numpy as np
from pathlib import Path
import simframe
v_sim = "1.0.6"
v_dpy = "1.0.9"
v_dpl = "1.0.6"
v_rad = "2.0"
msgs = []
if not simframe.__version__ == v_sim:
    msgs.append(f"Simframe version {v_sim} required, but version {simframe.__version__} found.")
if not dustpy.__version__ == v_dpy:
    msgs.append(f"DustPy version {v_dpy} required, but version {dustpy.__version__} found.")
if not dustpylib.__version__ == v_dpl:
    msgs.append(f"DustPyLib version {v_dpl} required, but version {dustpylib.__version__} found.")
! radmc3d > radmc3d.out
radmcfile = Path(".").joinpath("radmc3d.out")
with open(radmcfile, "r") as file:
    lines = file.readlines()
radmcfile.unlink()
if not np.any([f"VERSION {v_rad}" in l for l in lines]):
    msgs.append(f"RADMC-3D Version {v_rad} not found.")
def print_msg(msgs):
    if len(msgs)==0:
        m = "## Installation successful!"
    else:
        m = "## Installation unsuccessful:"
        for msg in msgs:
            m += "\n * "+msg
    return md(m)
```

```python
print_msg(msgs)
```
