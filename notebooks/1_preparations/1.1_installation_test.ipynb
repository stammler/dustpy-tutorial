# 1.1 Installation Test

Run this notebook to test if the installation of DustPy and DustPyLib has worked properly.  
The purpose of all commands will be explained in the tutorial notebooks.

```python
from dustpy import plot
from dustpy import Simulation
from dustpy import constants as c
from dustpylib.radtrans import radmc3d
import numpy as np
```

## 1.1.1 Testing DustPy

```python
sim = Simulation()
sim.initialize()
sim.t.snapshots = np.geomspace(1.e2, 1.e3, 2) * c.year
sim.writer.datadir = "test"
sim.run()
```

```python
plot.panel(sim)
```

## 1.1.2 Testing DustPyLib

```python
rt = radmc3d.Model(sim)
rt.datadir = "test_radmc3d"
rt.write_files()
```
